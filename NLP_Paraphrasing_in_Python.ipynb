```python
from parrot import Parrot
import torch
import warnings
warnings.filterwarnings("ignore")
```

```python
#Init models (make sure you init ONLY once if you integrate this to your code)
parrot = Parrot(model_tag="prithivida/parrot_paraphraser_on_T5", use_gpu=False)
```

```python
phrases = ["What are some good Biryani restaurants in Bangalore?"
]
```

```python
for phrase in phrases:
  print("-"*100)
  print("Input_phrase: ", phrase)
  print("-"*100)
  para_phrases = parrot.augment(input_phrase=phrase, do_diverse= False)
  for para_phrase in para_phrases:
   print(para_phrase)
```

```python
for phrase in phrases:
  print("-"*100)
  print("Input_phrase: ", phrase)
  print("-"*100)
  para_phrases = parrot.augment(input_phrase=phrase, do_diverse= True )
  for para_phrase in para_phrases:
   print(para_phrase)
```

```python
para_phrases = parrot.augment(input_phrase=phrase, 
                               diversity_ranker="levenshtein",
                               do_diverse=True, 
                               max_return_phrases = 10, 
                               max_length=32, 
                               adequacy_threshold = 0.2, 
                               fluency_threshold = 0.30)
```

```python
for para_phrase in para_phrases:
   print(para_phrase)
```
